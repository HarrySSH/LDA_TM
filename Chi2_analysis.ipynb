```python
def generate_freq_heatmap(Dataset_list = [], String_List = []):
    assert len(String_List) != 0, 'No input for words'
    assert len(Dataset_list) != 0, 'No input for Datasets'
    
    mtx = np.zeros((len(String_List), len(Dataset_list)))
    for i in range(len(Dataset_list)):
        for j in range(len(String_List)):
            length = Dataset_list[i].shape[0]
            mtx[j,i] = sum(Dataset_list[i]['note_text'].str.contains(String_List[j], na=False, regex=True))/length
            
    return mtx
```

```python
mtx = generate_freq_heatmap(Dataset_list, String_List)

sum_of_rows = mtx.sum(axis=1)
normalized_array = mtx / sum_of_rows[:, np.newaxis]
```

```python
'''
Lets tokenize each sentence into a list of words, removing punctuations and unnecessary characters altogether.
'''
# Convert to list
data = Dataset_list[0].note_text.values.tolist()
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(data))
```

```python


def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

# Remove Stop Words
data_words_nostops = remove_stopwords(data_words)
```

```python
def convert_to_sentence(_list):
    string = ''
    for x in _list:
        string +=' '+x
    return string
```
